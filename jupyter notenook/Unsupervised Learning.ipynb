# Title: Clustering Techniques on the Iris Dataset

## 1. Loading and Preprocessing

### Code


```python
# Import necessary libraries
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
iris_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Display the first few rows
iris_data.head()
```

# Explanation

The Iris dataset contains 150 samples of iris flowers with four features: sepal length, sepal width, petal length, and petal width. Since clustering is unsupervised, we drop the species column, which is not included in this dataset.


## 2. Clustering Algorithm Implementation

### A) KMeans Clustering

#### Description
KMeans clustering is an iterative algorithm that partitions a dataset into K distinct clusters based on feature similarity. It works by:

1. Initializing K centroids randomly.
2. Assigning each data point to the nearest centroid.
3. Recalculating centroids based on the mean of the assigned points.
4. Repeating the assignment and centroid calculation until convergence.

#### Suitability for the Iris Dataset
KMeans is suitable for the Iris dataset because:

- The dataset has clear groupings of the iris species based on the features.
- KMeans can handle large datasets and is efficient, making it ideal for the 150 samples in the Iris dataset.

## code:

```python
# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
iris_data['KMeans_Cluster'] = kmeans.fit_predict(iris_data)

# Visualization
plt.figure(figsize=(10, 6))
plt.scatter(iris_data.iloc[:, 0], iris_data.iloc[:, 1], c=iris_data['KMeans_Cluster'], cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('KMeans Clustering of Iris Dataset')
plt.show()
```

### B) Hierarchical Clustering

#### Description
Hierarchical clustering builds a tree of clusters (dendrogram) by either:

1. **Agglomerative method**: Starting with each data point as a single cluster and merging them based on similarity.
2. **Divisive method**: Starting with one cluster and recursively splitting it.

#### Suitability for the Iris Dataset
Hierarchical clustering is suitable for the Iris dataset because:

a) It provides a detailed view of the data structure through dendrograms.  
b) It does not require specifying the number of clusters in advance.

## Code:


```python
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage

# Hierarchical clustering
hierarchical = AgglomerativeClustering(n_clusters=3)
iris_data['Hierarchical_Cluster'] = hierarchical.fit_predict(iris_data)

# Visualization
plt.figure(figsize=(10, 6))
plt.scatter(iris_data['sepal length (cm)'], iris_data['sepal width (cm)'], c=iris_data['Hierarchical_Cluster'], cmap='plasma')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Hierarchical Clustering of Iris Dataset')
plt.show()
```
